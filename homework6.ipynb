```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
data = pd.read_csv("POP.csv")
```

```python
def my_mean(m):
    total = 0.0
    for x in m:
        total = total + float(x)
    return float(total)/len(m)
```

```python
def my_w1(X, Y):
    x_mean = my_mean(X)
    y_mean = my_mean(Y)
    num = 0.0
    den = 0.0
    for i in  range(len(X)):
        num = num + (X[i]- x_mean) * (Y[i] - y_mean)
        den = den + (X[i]- x_mean)*(X[i]- x_mean)
    return num/den
```

```python
def my_w0(X, Y):
    mean_y = my_mean(Y)
    mean_x = my_mean(X)
    return mean_y - my_w1(X, Y)*mean_x
```

```python
Y = data['value']

X = list(range(Y.size))
print(X)
print(Y)
```

```python
w1 = my_w1(X, Y)
w0 = my_w0(X, Y)
y_estimated= []
for x in X:
    y = w1 * x + w0
    y_estimated.append(y)
```

```python
x_samples = np.linspace(0, 1000, 1000)
y_samples = []
for x in x_samples:
    y_samples.append(w0+w1*x)
```

```python
plot.plot(X, Y, 'o')
plot.plot(x_samples, y_samples, color='green')
plot.xlabel('Index')
plot.ylabel('Value')
```

```python
plot.axis([0, 7, 0, 7])
plot.plot(X, Y, 'o')
plot.plot(x_samples, y_samples, color='green')
for i in range(len(X)):
    x_pair = [X[i], X[i]]
    y_pair = [Y[i], y_estimated[i]]
    plot.plot(x_pair, y_pair, color='red')
plot.xlabel('Index')
plot.ylabel('Value')
```
